```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

def func(x):
    return (x[0] - 2) ** 2 + (x[1] - 3) ** 2

x = np.linspace(0, 5, 100)
y = np.linspace(0, 6, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Строим поверхность
ax.plot_surface(X, Y, Z)

# Устанавливаем метки на осях
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')

# Выводим график
plt.title('График функции f(x, y) = x^2 + y^2')
plt.show()
```

```python
def func2(x):
    return x[0]**2 + x[1]**2 + x[0]*x[1]

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = func2([X, Y])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Строим поверхность
ax.plot_surface(X, Y, Z)

# Устанавливаем метки на осях
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')

# Выводим график
plt.title('График функции f(x, y) = x^2 + y^2 + xy')
plt.show()
```

```python
x0 = np.array([0, 0])
result = minimize(func, x0, method='Nelder-Mead')
print("Метод оптимизации нулевого порядка (Nelder-Mead):")
print("Результат:", result.x)
print("Значение функции:", result.fun)

result2 = minimize(func, x0, method='BFGS')
print("\nМетод оптимизации первого порядка (BFGS):")
print("Результат:", result2.x)
print("Значение функции:", result2.fun)
```
