```python
import numpy as np
import matplotlib.pyplot as plt

def dichotomy_method(a, b, func, eps):
    while b - a >= 2 * eps:
        c = (a + b) / 2 - eps
        d = (a + b) / 2 + eps
        if func(c) > func(d):
            a = c
        else:
            b = c
    return (a + b) / 2

def golden_section_search(a, b, func, eps):
    tao = (np.sqrt(5) - 1) / 2
    c = a + (1 - tao) * (b - a)
    d = b - (1 - tao) * (b - a)
    f_c = func(c)
    f_d = func(d)
    
    while b - a >= 2 * eps:
        if f_c < f_d:
            b = d
            d = c
            c = a + (1 - tao) * (b - a)
            f_d = f_c
            f_c = func(c)
        else:
            a = c
            c = d
            d = b - (1 - tao) * (b - a)
            f_c = f_d
            f_d = func(d)
    return (a + b) / 2

def newton_method(func, derivative, x0, eps, max_iter):
    x = x0
    for i in range(max_iter):
        x = x - func(x) / derivative(x)
        if abs(func(x)) < eps:
            return x
    return x

def func(x):
    return (x - 2) ** 2

def func_derivative(x):
    return 2 * (x - 2)

x = np.linspace(0, 4, 400)
y = func(x)
plt.plot(x, y, label='f(x) = (x - 2)^2')

#dichotomy method:
result_binary = dichotomy_method(0, 4, func, 0.0001)
print("dichotomy method result:", result_binary)

plt.scatter(result_binary, func(result_binary), color='red', label='dichotomy method')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
plt.plot(x, y, label='f(x) = (x - 2)^2')

#golden section search:
result_golden = golden_section_search(0, 4, func, 0.0001)
print("golden section search result:", result_golden)

plt.scatter(result_golden, func(result_golden), color='purple', label='golden section search')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
plt.plot(x, y, label='f(x) = (x - 2)^2')

#newton method:
result_newton = newton_method(func, func_derivative, 1, 0.0001, 100)
print("newton result:", result_newton)

plt.scatter(result_newton, func(result_newton), color='green', label='newton method')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
